# 13.원 그래프 (기본)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# values = [1,1,1,1,1,1]
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%') # labels 라벨에 labels 값을 넣어준다 / autopct='#.1f'로 숫자 값 넣어주기
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) # 90도 로 차트 시작하기, 시계 방향으로 정렬하기
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
# explode=[0.2, 0.1, 0, 0, 0, 0] # 파이썬은 0.2만큼, 자바는 0.1 만큼 띄우고 나머지는 그대로 두기
explode =[0.05] * 6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```
