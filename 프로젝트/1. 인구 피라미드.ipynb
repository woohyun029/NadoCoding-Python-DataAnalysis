## 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 0 자리의 row의 문자열에서 ,를 빈 문자열로 바꾼 뒤 int(정수형)으로 타입 변경
df_m
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # 컬럼 명 통일하기 (표를 같이 그리기 위해서)
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w
```

## 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2011 대한민국 인구 피라미드')
plt.savefig('2011_인구피라미드.png', dpi=100)
plt.show()
```

```python
file_name = '202601_202601_연령별인구현황_월간.xlsx'
df_m_1 = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_w_1 = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
plt.figure(figsize=(10,7))
df_m_1.iloc[0] = df_m_1.iloc[0].str.replace(',', '').astype(int)
df_w_1.columns = df_m_1.columns
df_w_1.iloc[0] = df_w_1.iloc[0].str.replace(',', '').astype(int)
plt.figure(figsize=(10,7))
plt.barh(df_m_1.columns, -df_m_1.iloc[0] // 1000) # 단위 : 천 명
plt.barh(df_w_1.columns, df_w_1.iloc[0] // 1000)
plt.title('2021 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png', dpi=100)
plt.show()
```
